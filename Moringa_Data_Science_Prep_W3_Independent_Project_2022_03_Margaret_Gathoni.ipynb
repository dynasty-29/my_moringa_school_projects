We will now go ahead and combine the three dataset in to one for further analysis

```python
## combining the three dataset
tel_data = pd.concat([data_1, data_2, data_3], axis=0)
tel_data.head(5)
```

```python
#From the instructions given the country columns are not required
# so we go on and Drop Not required Columns
teldata = tel_data.drop(['COUNTRY_A', 'COUNTRY_B'], axis=1)

#WE go on and view to see the drop fucntion in action
teldata.head(4)
```

```python
teldata.info()
```

```python
#change date_time datatype to datetime64 
teldata['DATE_TIME'] = pd.to_datetime(tel_data['DATE_TIME'])
teldata.info()
```

```python
## WE now go ahead and drop duplicates
teldat = teldata.drop_duplicates()
teldat.info()

```

```python
## review the dataset after dropping duplicates
teldat.head()
```

```python
#sorting using date.
teldat=teldat.sort_values(by=["DATE_TIME"],ascending=True)
teldat.head()
```

```python
# We want to find out the Cities which was more used for the three days
# so will go ahead and join the teldat to the cells_geo data to have all the values in question 
#we want to join the two keeping all the rows and columns for both tables so we use outer joins
# we will merge with a common column cell_id

tel_cells = pd.merge(teldat,data, how='outer', on=['CELL_ID'])
```

```python
tel_cells.shape
```

1. Which ones were the most used city for the three days?
The most used city should have the most calls hence will filter using cellid and the product column

```python
#To get the most used city we will check for the city with the most calls using the cell_id column
#while sorting with the product column which stores sms and voice calls
tel_cells.groupby('CELL_ID').count().nlargest(5,'PRODUCT')
```

WE can see that 'ffa6759bb2' is the most used cell_id

2. Which cities were the most used during business and home hours?

To answer this question will try to understand what it mean
most used city during business and home hours means the product value should be maximum then we filter by city



```python
#will go ahead and join these two tables first using the left join
#further filter using city on the largest value after joining
tel_cells = pd.merge(data,teldat, how='left', on=['CELL_ID'])
tel_cells
tel_cells.groupby('CITY').count().nlargest(5,'VALUE')
```

We can see that Abengourou, Abobo, Aboisso were the top three cities busiest cities both in business and home hours
